# Mengubah Fitur-Fitur Bertipe Datetime Menjadi Numerik

Memodifikasi fitur-fitur bertipe date dari format bulan-tahun menjadi perhitungan berapa lama waktu berlalu sejak waktu tersebut. Untuk melakukan hal ini, umumnya digunakan reference date = hari ini. Namun, karena dataset ini merupakan dataset tahun 2007-2014, maka akan lebih relevan jika menggunakan reference date di sekitar tahun 2017. Dalam contoh ini, saya menggunakan tanggal 2017-12-01 sebagai reference date.

```python
new_cols = [(col+'_month') for col in datetime_features]
new_cols
```

```python
df.shape
```

```python
for col in datetime_features:
    df[col+'_month'] = round(
        pd.to_numeric((pd.to_datetime('2017-12-01') - df[col]) / np.timedelta64(1, 'M'))
    )
df.shape
```

```python
df[new_cols].head(5)
```

pada kolom 'earliest_cr_line_month' nilai min bernilai minus yang mana hal ini tidak masuk akal, periksa penyebab dari munculnya nilai minus tersebut

```python
df.loc[
    df['earliest_cr_line_month'] < 0, 
    np.concatenate((new_cols,datetime_features))
]
```

terdapat 1169 yang memiliki nilai minus pada kolom 'earliest_cr_line_month', ternyata hal ini disebabkan oleh nilai tahun yang tidak wajar pada kolom 'earliest_cr_line' yaitu tahun yang jauh melewati tahun yang dijadikan reference yaitu 2017. Ini terjadi karena program mengkonversi 2 digit terakhir seperti 64 yang seharusnya berarti 1964 menjadi 2064. 

Selanjutnya kita akan meng-handle nilai yang tidak wajar ini dengan merubah 2 digit pertama tahunya menjadi '19' bukan '20'

```python
def handle_unrealistic_year(datetime_obj):
    temp = ''
    new_year = 0
    year = datetime_obj.year
    if  year > 2017:
        temp = str(year)[2:]
        new_year = int('19'+temp)
        datetime_obj = datetime_obj.replace(year=new_year)
    return datetime_obj
```

```python
df.loc[df['earliest_cr_line'] > '2017-12-01',['earliest_cr_line']].head(5)
```

```python
df.loc[df['earliest_cr_line'].dt.year > 2017 ,['earliest_cr_line']].head(5)
```

```python
df['earliest_cr_line'] = df['earliest_cr_line'].map(handle_unrealistic_year)
```

```python
df.loc[df['earliest_cr_line'] > '2017-12-01',['earliest_cr_line']]
```

# Standarisasi

Algoritma machine learning memiliki performa lebih baik dan konvergen lebih cepat ketika dimodelkan pada data dengan skala relatif sama atau mendekati distribusi normal. Proses scaling dan standarisasi membantu untuk membuat fitur data menjadi bentuk yang lebih mudah diolah oleh algoritma.

Standardisasi adalah teknik transformasi yang paling umum digunakan dalam tahap persiapan pemodelan. Untuk fitur numerik, kita tidak akan melakukan transformasi dengan one-hot-encoding seperti pada fitur kategori. Kita akan menggunakan teknik StandarScaler dari library Scikitlearn,

StandardScaler melakukan proses standarisasi fitur dengan mengurangkan mean (nilai rata-rata) kemudian membaginya dengan standar deviasi untuk menggeser distribusi. StandardScaler menghasilkan distribusi dengan standar deviasi sama dengan 1 dan mean sama dengan 0. Sekitar 68% dari nilai akan berada di antara -1 dan 1.

Untuk menghindari kebocoran informasi pada data uji, kita hanya akan menerapkan fitur standarisasi pada data latih. Kemudian, pada tahap evaluasi, kita akan melakukan standarisasi pada data uji. Untuk lebih jelasnya, mari kita terapkan StandardScaler pada data.
